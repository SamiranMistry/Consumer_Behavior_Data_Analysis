## Solving the missing values for Review Rating

#### *Using Median for each product categories to fill the missing values*

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

## Formatting the column names
#### *Replacing all the space with underscore '_'*
#### *Lowercasing all the letters*
by doing this we can minimise the chances or error

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

#### Creating a column age_group

```python
labels=['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

#### Creating column purchase_frequency_days

```python
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90,
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
